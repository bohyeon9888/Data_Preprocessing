```python
with open(file, 'r') as json_file:
    data = json.load(json_file)
```

```python
# 합쳐진 데이터, json 파싱 단계
for filename, file_data in data.items():

    tester = filename

    for list in file_data:

        pattern = list.get("pattern")[0]
        pattern = pattern.strip('[]')
        # print("Pattern : ", list.get("pattern"))

        matchesArr = list.get("matches")
        for match in matchesArr:

            onTime = match.get("ontime")
            offTime = match.get("offtime")
            userPresent = match.get("userpresent")

            screen_ontime = abs(onTime - offTime)
            screenUnlocking = 0
            if (userPresent != None):
                screenUnlocking = userPresent - onTime

            eventsArr = match.get("events")

            for event in eventsArr:
                patternIdx = event.get("patternIndex")
                dataType = event.get("datumType")
                timeStamp = event.get("timestamp")
                deviceType = event.get("deviceType")
                package = event.get("packageName")
                classN = event.get("className")
                currentKey = event.get("currentKey")
                timetaken = event.get("timeTaken")
                newRow = {'tester': tester, 'pattern': pattern, 'screenOn': onTime, 'screenOff': offTime, 'screen on/off': screen_ontime,
                          'screen unlocking': screenUnlocking, 'pattern Idx': patternIdx, 'Data Type': dataType, 'Timestamp': timeStamp,
                          'Device Type': deviceType, 'Package Name': package, 'Class Name': classN, 'currentKey': currentKey, 'Timetaken': timetaken}

                tmpdf = tmpdf.append(newRow, ignore_index=True)
```

```python
# timestamp 및 pattern index 정수형으로 보이게 수
tmpdf['screenOn'] = tmpdf['screenOn'].astype('int64')
tmpdf['screenOff'] = tmpdf['screenOff'].astype('int64')
tmpdf['Timestamp'] = tmpdf['Timestamp'].astype('int64')
tmpdf['pattern Idx'] = tmpdf['pattern Idx'].astype(int)
```

```python
# 각 pattern에 대해서 업무 수행할 수 있도록 각 pattern의 배열을 만듦
pattern_arr = tmpdf['pattern'].unique()
```

```python
for i in range(len(pattern_arr)):
    bufferdf = tmpdf[(tmpdf['pattern']==pattern_arr[i])]
    tester = bufferdf.iloc[0, 0]

    # number of event frequency
    bufferdf['package_count'] = bufferdf.groupby(['Package Name', 'pattern Idx'])['Package Name'].transform('count')

    # task completion time - start, end
    bufferdf['start'] = bufferdf.groupby('pattern Idx')['Timestamp'].transform('first')
    bufferdf['end'] = bufferdf.groupby('pattern Idx')['Timestamp'].transform('last')

    # notification response time
    notidf = bufferdf[(bufferdf['Data Type']=='NOTIFICATION')]
    if not notidf.empty:
      noti_file = f'/content/{tester}.csv'
      tester_df = pd.read_csv(noti_file)

      for index1, row1 in notidf.iterrows():
        package_name = row1['Package Name']
        timestamp1 = row1['Timestamp']

        filtered = tester_df[(tester_df['packageName'] == package_name) & (tester_df['timestamp'] > timestamp1)]

        if not filtered.empty:
          first_row = filtered.iloc[0]
          timestamp2 = first_row['timestamp']

          response_time = timestamp2 - timestamp1
          notidf.at[index1, 'Noti ResponseTime'] = response_time

    # Package Name 및 Pattern Index별로 같은 값들을 모아 평균을 냄
    bufferdf['noti_response'] = notidf.groupby(['Package Name', 'pattern Idx'])['Noti ResponseTime'].transform('mean')


    # typing metric
    typingdf = bufferdf[(bufferdf['Device Type']=='typeViewTextChanged')]
    if typingdf.empty:
        bufferdf = bufferdf.drop_duplicates(subset=['pattern Idx', 'Device Type', 'Package Name', 'Class Name'])
        for index, row in bufferdf.iterrows():
            newRow = {'tester': row['tester'], 'Pattern': row['pattern'], 'Screen On': row['screenOn'],  'Screen Off': row['screenOff'],
                      'Screen On/Off Time': row['screen on/off'],
                      'Screen Unlocking Time': row['screen unlocking'], 'Pattern Idx': row['pattern Idx'], 'Data Type': row['Data Type'],
                      'Device Type': row['Device Type'], 'Package Name': row['Package Name'], 'Class Name': row['Class Name'],
                      'Task Completion-start': row['start'], 'Task Completion-end': row['end'], 'Event Frequency': row['package_count'],
                      'Noti ResponseTime': row['noti_response']}
            outputdf = outputdf.append(newRow, ignore_index=True)
    else:
        newdf = pd.DataFrame(columns=typingdf.columns)

        for pattern_idx, group in typingdf.groupby('pattern Idx'):
            newdf = group.copy()

            inter_arr = newdf.loc[:, 'Timetaken']
            last_index = newdf.tail(1).index[0]
            inter_arr = inter_arr.reset_index(drop=True)
            if last_index + 1 in bufferdf.index:
                last_time = bufferdf.loc[last_index+1, 'Timestamp'] - bufferdf.loc[last_index, 'Timestamp']
                new_series = pd.Series([last_time])
                intra_arr = inter_arr.append(new_series, ignore_index=True)
            else:
                intra_arr = inter_arr

            inter_arr = inter_arr.drop(0, axis=0)
            intra_arr = intra_arr.drop(0, axis=0)

            countIS = len(newdf)
            backCnt = (newdf['currentKey'] == '\u0008').sum()

            word_T = countIS - backCnt #한 view의 typing 개수에서 backspace를 뺀 것
            first_s=newdf.iloc[0, 8] #typing 을 시작한 지점의 timestamp
            last_s=newdf.iloc[-1, 8] #typing 을 종료한 지점의 timestamp
            time_dif = round(((last_s - first_s)/1000), 1) #typing을 한 총 시간 (minute 단위)

            if word_T != 0 and time_dif != 0:
                CPS = (word_T-1) / time_dif
                CPM = CPS * 60
                WPS = CPS / 5
                WPM = WPS * 60
                KSPS = countIS / time_dif
                KSPC = countIS / word_T
            elif word_T == 0 and time_dif != 0:
                CPS = CPM = WPS = WPM = KSPC = 0
                KSPS = countIS / time_dif
            elif word_T != 0 and time_dif == 0:
              CPS = CPM = WPS = WPM = KSPS = 0
              KSPC = countIS / word_T
            else:
              CPS = CPM = WPS = WPM = KSPS = KSPC = 0

            Inter = (inter_arr.sum()) / countIS - 1
            Intra = (intra_arr.sum()) / countIS
            newdf = newdf.reset_index(drop=True)

            newRow = {'tester': tester, 'Pattern': newdf.loc[0, 'pattern'], 'Screen On': newdf.loc[0, 'screenOn'],  'Screen Off': newdf.loc[0, 'screenOff'],
                      'Screen On/Off Time': newdf.loc[0, 'screen on/off'], 'Screen Unlocking Time': newdf.loc[0, 'screen unlocking'],
                      'Pattern Idx': newdf.loc[0, 'pattern Idx'], 'Data Type': newdf.loc[0, 'Data Type'], 'Device Type': newdf.loc[0, 'Device Type'],
                      'Package Name': newdf.loc[0, 'Package Name'], 'Class Name': newdf.loc[0, 'Class Name'],
                      'Task Completion-start': newdf.loc[0, 'start'], 'Task Completion-end': newdf.loc[0, 'end'],
                      'Event Frequency': newdf.loc[0, 'package_count'], 'Noti ResponseTime': row['noti_response'], 'CPS': CPS, 'CPM': CPM,
                      'WPS': WPS, 'WPM': WPM, 'KSPS': KSPS,
                      'KSPC': KSPC, 'Inter': Inter, 'Intra': Intra}
            outputdf = outputdf.append(newRow, ignore_index=True)
```
